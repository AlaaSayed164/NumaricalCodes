#### Use your code from practical session 1 and create a function to perform GD for single variable linear regression.
#### The function should take the following input parameters:
##### Input data (X), Target Labels (y), Learning Rate (alpha), Maximum number of iterations
#### The function should return the following:
##### All Theta_0 and Theta_1 values till converge, and equivellant Losses, all hypothesis output, and optimum values of Theta_0 and Theta_1
#### Note: you can choose any suitable stop criteria.

```python
def GD_Batch(x_points,y_points,alpha,max_iterations):
    x_points= np.array(x_points)
    y_points= np.array(y_points)
    theta_0=np.array([0])
    theta_1=np.array([0])
    theta_0s= [theta_0]
    theta_1s= [theta_1]
    MSEs=[10]
    Hs=[]
    j=1000
    iterations = 0
    while (j>= 0.001)and (iterations <max_iterations) :
        iterations+=1
        y_predicted= theta_0+theta_1*x_points
        Hs.append(y_predicted)
        j_theta_0=  (np.subtract(y_predicted,y_points)).mean()
        j_theta_1=  (x_points*(np.subtract(y_predicted,y_points))).mean()
        theta_0 = theta_0 - (alpha*j_theta_0)
        theta_0s.append(theta_0)
        theta_1 = theta_1 - (alpha*j_theta_1)
        theta_1s.append(theta_1)
        j= np.sqrt(np.square(j_theta_0)+ np.square(j_theta_1))
        MSE = 0.5*np.square(np.subtract(y_predicted,y_points)).mean()
        MSEs.append(MSE)
        if abs(MSEs[-2]-MSEs[-1])< 0.000001:
            break
        #print(j)
    print("Optimal Theta_0 = " ,theta_0)
    print("Optimal Theta_1 = " ,theta_1)
    print("Number of epochs = ", iterations)
  #  print("All Theta_0 values: ", theta_0s)
  # print("All Theta_1 values: ", theta_1s)
  #  print("All hypothesis : ",Hs)
  # print("Losses: ", MSEs)
    print("R2_score = ",r2_score(y_points, y_predicted))
    return (theta_0,theta_1,theta_0s,theta_1s,Hs, MSEs)
```

#### Update the implementation of the Batch GD to be Mini-Batch and rpeat the above plots and evaluations
#### Change batch size and see the effect

```python
def GD_MiniBatch(x_points,y_points,batch_size,alpha,max_iterations):
    b= len(x_points)// batch_size
    x_points= np.split(np.array(x_points),b)
    y_points= np.split(np.array(y_points),b)
    theta_0=np.array([0])
    theta_1=np.array([0])
    theta_0s= [theta_0]
    theta_1s= [theta_1]
    MSEs=[10]
    Hs=[]
    j=1000
    iterations = 0
    while (j>= 0.001)and (iterations <max_iterations) :
        iterations+=1
        for i in range (b):
            y_predicted= theta_0+theta_1*x_points[i]
            Hs.append(y_predicted)
            j_theta_0=  (np.subtract(y_predicted,y_points[i])).mean()
            j_theta_1=  (x_points[i]*(np.subtract(y_predicted,y_points[i]))).mean()
            theta_0 = theta_0 - (alpha*j_theta_0)
            theta_0s.append(theta_0)
            theta_1 = theta_1 - (alpha*j_theta_1)
            theta_1s.append(theta_1)
            j= np.sqrt(np.square(j_theta_0)+ np.square(j_theta_1))
            MSE = 0.5*np.square(np.subtract(y_predicted,y_points[i])).mean()
            MSEs.append(MSE)
            if abs(MSEs[-2]-MSEs[-1])< 0.000001:
                break
        #print(j)
    print("Optimal Theta_0 = " ,theta_0)
    print("Optimal Theta_1 = " ,theta_1)
    print("Number of epochs = ", iterations)
  #  print("All Theta_0 values: ", theta_0s)
  # print("All Theta_1 values: ", theta_1s)
  #  print("All hypothesis : ",Hs)
  # print("Losses: ", MSEs)
  # print("R2_score = ",r2_score(y_points, y_predicted))
    return (theta_0,theta_1,theta_0s,theta_1s,Hs, MSEs)
```

```python
Res=GD_MiniBatch(x_points,y_points,4,0.01,1000)
```

```python
y_pred= Res[0]+ Res[1]*x_points
print("R2_score = ",r2_score(y_points, y_pred))
```
